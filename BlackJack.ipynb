```python
def Blackjack():
    number_of_players=input("How many people will play?\n")
    try:
        number_of_players=int(number_of_players)
    except:
        number_of_players=0
    while number_of_players<=0:
        print("Are there really no players? Let me ask again.")
        number_of_players=input("How many people will play?\n")
        try:
            number_of_players=int(number_of_players)
        except:
            number_of_players=0
    players=[]
    for player_names in range(number_of_players):   #Creates a list with the names of the players
        players.append(input(f"Insert Player {player_names+1}\'s name.\n"))
    active_players=[True for player in players]     #This list is used to define who is still playing
    score_players=[0 for player in players]     #This list will save the scores of the players
    score1=[]
    score2=[]
    score_total=[]
    def Create_Deck():
        blackjack_deck=[]
        score_cards=[]
        cards=["Ace","Two","Three","Four","Five","Six","Seven","Eight","Nine","Ten","Jack","Queen","King"]
        suits=["Diamonds","Spades","Hearts","Clubs"]
        suits_count=0
        for suits_count in range(4):
            for cards_count in range(13):
                blackjack_deck.append(f"{cards[cards_count]} of {suits[suits_count]}")
        return blackjack_deck
    def Play():
        score_total=[0,[]]
        while True in active_players:   #Checks the list for active players
            for player in range(len(players)):  #runs the list of players
                if active_players[player]==True:    #from the list of players, we select the next active player           
                    score_total=[0,[]]  #resetting this variable
                    while score_players[player]<21 and active_players[player]==True:    #while the player is still active / lose condition
                        print(f"{players[player]}, you have {score_players[player]} points.")    #shows the player the score before drawing a card
                        draw_card=input(f"{players[player]}, do you wish to draw a card? YES or NO?\n")
                        while draw_card!="YES" and draw_card!="NO":    #verifies the input
                            print("Answer with YES or NO.")
                            draw_card=input(f"{players[player]}, do you wish to draw a card? YES or NO?\n")
                        if draw_card=="YES":
                            score1=Draw()    #receives a list with [total score , [cards drawn]]
                            no_repetition=0
                            while no_repetition==0:   #checks if the player already drew that card to avoid repetition
                                if score1[1] in score_total[1]:    #if the card received is in the list of cards the player has already drawn
                                    score1=Draw()   #draw again
                                else:
                                    no_repetition=1
                            score_total[0]=score_total[0]+score1[0]
                            score_total[1].extend(score1[1])
                            if score_total[0]==21:
                                print(f"You drew {score_total[1]} and summed 21 points!\nBlackjack!\nCongratulations!\nNext player.")
                                score_players[player]=21
                                active_players[player]="NO"
                            elif score_total[0]>21:
                                print(f"You drew {score1[1]} and summed {score1[0]} more points.\n",
                                f"Your total is {score_total[0]} and you drew {score_total[1]}.\n",
                                f"You Lose!\nYour total score will be 0.")
                                
                                active_players[player]="NO"
                            elif score_total[0]<21:
                                print(f"You drew {score1[1]} and summed {score1[0]} more points.\n",
                                f"Your total is {score_total[0]} and you drew {score_total[1]}.")
                                score_players[player]=score_total[0]
                        elif draw_card=="NO":
                            print("Ok. Next player.")
                            active_players[player]="NO"
                            
                        else:
                            while draw_card!="YES" and draw_card!="NO": #Verifies the input
                                print("Answer with YES or NO.")
                                draw_card=input("Do you with to draw a card? YES or NO?\n")
                else:
                    print(f"There are {active_players.count(True)} active players.")
    def Draw():
        import random
        card=random.randint(0,51)
        deck=Create_Deck()
        card_drawn=deck[card]
        list_draw=[0,[]]                                                   
        cards_one_point=["Ace"]
        cards_numbers=["Two","Three","Four","Five","Six","Seven","Eight","Nine","Ten"]
        cards_faces=["Jack","Queen","King"]                                                    
        for element in cards_one_point:
            if element in card_drawn:
                list_draw[0]=1
                list_draw[1].append(card_drawn)
        for element in cards_numbers:
            if element in card_drawn:
                list_draw[0]=(cards_numbers.index(element))+2
                list_draw[1].append(card_drawn)
        for element in cards_faces:
            if element in card_drawn:
                list_draw[0]=10
                list_draw[1].append(card_drawn)
        return list_draw
    def Verify():
        winner_score=max(score_players)     #determine the highest score
        winners=[]
        winners_return=[]
        for element in range(len(score_players)):
            if score_players[element]==winner_score:   #this is useful when more than one player has the highest score
                 winners.append(element)
        winners_return=[players[i] for i in winners]
        print(f"The player(s) with the highest score is(are): {winners_return} with {winner_score} points.")
        print(f"Congratulations, {winners_return}!")
    Play()
    Verify()
    
```
